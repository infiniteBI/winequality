```python
# importing the data df1 white wine, df2 red wine
df1 = pd.read_csv('C:/Users/JP laptop/Documents/winequality/winequalitywhite.csv')
df2 = pd.read_csv('C:/Users/JP laptop/Documents/winequality/winequalityred.csv')
```

```python
#let's check for missing data
for col in df1.columns:
    pct1_missing = np.mean(df1[col].isnull())
    print('{} - {}%'.format(col, round(pct1_missing*100)))
```

```python
#let's see the top wine quality
df1.sort_values(by=['quality'],inplace=False, ascending= False).head(5)
```

```python
#let's see if there is correlation
correlation_matrix1 = df1.corr()
sns.heatmap(correlation_matrix1, annot = True)
plt.show()
```

```python
#Let's see the average values base on quality
df1.groupby(['quality']).mean().sort_values(by=['quality'],inplace=False, ascending= False)
```

```python
#let's see top correlation for white wine
top_white_corr = correlation_matrix1.unstack()
top_white_corr_sorted= top_white_corr.sort_values(inplace = False, ascending= False) 
top_white_corr_strong= top_white_corr_sorted[top_white_corr_sorted >= 0.30]
top_white_corr_strong= top_white_corr_strong[top_white_corr_strong < 1]
print(top_white_corr_strong)
```

```python
#let's see top correlation for red wine
top_white_corr1 = correlation_matrix2.unstack()
top_white_corr_sorted1= top_white_corr1.sort_values(inplace = False, ascending= False) 
top_white_corr_strong1= top_white_corr_sorted1[top_white_corr_sorted1 >= 0.30]
top_white_corr_strong1= top_white_corr_strong1[top_white_corr_strong1 < 1]
print(top_white_corr_strong1)
```

```python
#Multiple linear regression
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X= df1.drop(columns= 'quality')
y= df1['quality']
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size  =0.3, random_state=0)
```

```python
lr= LinearRegression()
lr.fit(X_train,y_train)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
y_pred_train= lr.predict(X_train)
y_pred_train
```

```python
plt.scatter(y_train, y_pred_train)
plt.xlabel('y_train')
plt.ylabel('Prediction')
plt.show()
```

```python
y_pred_test= lr.predict(X_test)
y_pred_test
```

```python
plt.scatter(y_test, y_pred_test)
plt.xlabel('y_test')
plt.ylabel('Test Prediction')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred_test)
r2
```
